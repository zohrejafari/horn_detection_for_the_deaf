```python
import tensorflow as tf
from keras.models import Model
from keras.layers import Input, LSTM, Dense,Dropout
import numpy as np
import keras as keras
from keras import regularizers
from keras.models import Sequential
from keras.optimizers import *
import matplotlib.pyplot as plt
from scipy.io import wavfile
from keras import metrics
import librosa
import math
import glob
import os
import sys
from keras.utils import to_categorical
from sklearn.metrics import mean_squared_error
from sklearn import preprocessing
from keras.callbacks import LambdaCallback
import keras.backend as K
import datetime
from keras.callbacks import EarlyStopping, ModelCheckpoint
import random
```

```python
class DATA():
    def __init__(self,car_horn_file_name,street_voice_file_name):
        
        self._carHorn=car_horn_file_name
        self._street=street_voice_file_name
        
        self._voice=np.zeros((4000,3001),np.float32)
        
        self._num=0
        self._extractData(self._carHorn,1) 
        self._extractData(self._street,0)
        self._seprateData()
        
    def _extractData(self,filename,label):
        
        for fn in glob.glob(os.path.join(filename, "*.wav")):
            try:
                print('Extraccting data for ', fn)
                Y, sample_rate = librosa.load(fn,1000)
                x=Y.shape
               
                if x[0]>3000:
                    Y=Y[0:3000]       
                    
                else:
                    temp=np.zeros((3000))
                    for i in range(0,3000):
                        temp[i]=Y[i%x[0]]
     
                    Y=temp
                  
                    
                    
                outputarray = self._extractFeature(Y) 
                output=self._makeLabel(outputarray,label)
                
                self._voice[self._num]=output
                self._num=self._num+1
                print(self._num)
                
            except Exception as e:
                print ("Error encountered while parsing file: ", fn)
                print(e)
                continue
                
          
        
        
        
    
    def _makeLabel(self,array,label):
        temp=np.append(array, label)
        return temp
    
    def _seprateData(self):
        random.Random(datetime.datetime.now()).shuffle(self._voice)
        np.random.shuffle(self._voice)
        label=self._voice[:,3000:3001]
        data=self._voice[:,0:3000]
    
        label=to_categorical(label,2)
        
        self._trainLabel=label[0:2800,]
        self._valLabel=label[2800:3400,]
        self._testLabel=label[3400:4000,]
       
        pdata=preprocessing.scale(data)
        self._trainData=pdata[0:2800,]
        self._valData=pdata[2800:3400,]
        self._testData=pdata[3400:4000,]
    
   
        
    def _extractFeature(self,time_series):
        return time_series

    
    
    
    def getTrainData(self):
        return self._trainData,self._trainLabel
    
    def getTestData(self):
        return self._testData,self._testLabel
    
    def getValData(self):
        return self._valData,self._valLabel
    
```

```python

print('Loading data...')
trainData = np.load('train_data.npy')
trainLabel = np.load('train_label.npy')
valData = np.load('val_data.npy')
valLabel = np.load('val_label.npy')
testData = np.load('test_data.npy')
testLabel = np.load('test_label.npy')


```

```python
trainData=preprocessing.normalize(trainData,norm='l2')
valData=preprocessing.normalize(valData,norm='l2')
testData=preprocessing.normalize(testData,norm='l2')
```

```python
print('Training...')
```

```python
# create model
model = Sequential()
model.add(Dense(50, input_dim=3000, activation='relu',kernel_regularizer=regularizers.l2(0.001)))
model.add(Dropout(0.5))
model.add(Dense(20, activation='relu'))
model.add(Dense(2, activation='softmax'))
# Compile model
op=keras.optimizers.SGD(lr=0.1, momentum=0.9, decay=0.1/100, nesterov=False)
model.compile(loss='categorical_crossentropy', optimizer=op, metrics=['accuracy'])
```

```python
callbacks = [EarlyStopping(monitor='val_loss', patience=2), ModelCheckpoint(filepath='best_model.h5', monitor='val_loss', save_best_only=True)]

history = model.fit(trainData, trainLabel, epochs=100,batch_size=50,validation_data=(valData, valLabel), shuffle=True)
```

```python
print(testLabel)
yhat=model.predict_classes(testData)
print(yhat)
```
